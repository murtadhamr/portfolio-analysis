# PTBA Stock Analysis

```python
ptba = yf.Ticker("PTBA.JK")

ptba_hist = ptba.history(period = 'max')
```

```python
ptba_hist['Datetime'] = ptba_hist.index
ptba_hist['Date'] = ptba_hist['Datetime'].dt.date
```

```python
ptba_hist.set_index('Date')
ptba_hist.info()
ptba_hist['Year'] = ptba_hist['Datetime'].dt.year
```

```python
ptba_chart = ptba_hist[ptba_hist['Datetime'] >= '2018-01-01']

plt.figure(figsize = (10, 6))

sns.lineplot(data = ptba_chart, x = 'Date', y = 'Open', color = 'red', label = 'Open')

sns.lineplot(data = ptba_chart, x = 'Date', y = 'Close', color = 'blue', label = 'Close')
plt.ylabel('Price')

plt.axhline(y = ptba_chart['Close'].mean(), color = 'green', linestyle = '--')


plt.xlabel('Year')
plt.title('PTBA Stock Price')
plt.legend()
```
